```python
question1
Values:

'hello' (string)
-87.8 (floating-point number)
6 (integer)
Expressions (Mathematical Operators):

(multiplication)
(subtraction/negation)
/ (division)
(addition)



question2
String:
A string is a data type used to represent a sequence of characters, such as letters, digits, symbols, and spaces.
It is used to store textual information. In programming, strings are usually enclosed within single ('') or double ("") 
quotation marks. For example:
"Hello, World!"

A variable is a named storage location in a program's memory that holds a value. It is used to store data that can be used,
 manipulated, and changed during the program's execution. Variables provide a way to refer to data by a meaningful 
name rather than its memory address.
name = "Alice"  # Here, "name" is a variable that holds a string value "Alice".
age = 25       # Here, "age" is a variable that holds an integer value 25.


question3
Integer (int):
An integer data type represents whole numbers without any fractional or decimal part. Integers can be positive, negative,
 or zero. They are used for counting, indexing, and performing arithmetic operations that involve whole numbers
age = 25

(float):
A floating-point data type, often referred to as a "float," is used to represent numbers with both an integral and a 
fractional part. Floating-point numbers are used when precision in representing real numbers is required.
pi = 3.14159

Boolean (bool):
A boolean data type has only two possible values: True and False. Booleans are used for representing binary decisions or
 states, such as the result of a comparison or a condition's evaluation. They are fundamental for controlling program
 flow through conditional statements and loops.


question4
An expression in programming is made up of one or more elements that combine to produce a value. Expressions are used to
 compute values or perform operations and are a fundamental building block of programming languages. Here are the component
s that make up expressions:

Operators: These are symbols that represent specific operations. For example, + is the addition operator, - is the
 subtraction operator, * is the multiplication operator, / is the division operator, and so on.

Operands: These are the values or variables that the operators act upon. For example, in the expression 3 + 5, 3 and 5
 are operands, and the + operator performs the addition operation on them.


queston5
Expression:
An expression is a combination of values, variables, operators, and function calls that can be evaluated to produce a 
single value. In other words, an expression computes or calculates a value. Expressions can be used as parts of larger 
expressions, assigned to variables, or used in various contexts where a value is expected.
Examples of expressions:

3 + 5 (evaluates to 8)
x * y (evaluates to the product of variables x and y)
len("hello") (evaluates to the length of the string "hello")

A statement is a complete unit of code that performs an action. Statements are used to control the flow of a program,
 perform assignments, define functions, and more. Unlike expressions, statements do not produce a value that can be used 
further in calculations. Instead, they are executed for their side effects or control flow.
Examples of statements:

Assignment statement: spam = 10 (assigns the value 10 to the variable spam)
Conditional statement: if x > 0: ... (controls program flow based on a condition)
Loop statement: for item in list: ... (repeats a block of code for each item in a list)
Function definition: def my_function(): ... (defines a named block of code that can be called later)

```
